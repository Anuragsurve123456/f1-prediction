```python
import pandas as pd
import numpy as np
import os

Data_path = "../Dataset/"
datasets = {}
for i in os.listdir(Data_path):
    if i.endswith(".csv"):
        datasets[i.replace(".csv","")] = pd.read_csv(os.path.join(Data_path, i))
```

```python
print(datasets)
```

```python
lap_times = datasets["lap_times"].copy()
qualifying = datasets["qualifying"].copy()

```

```python
def to_seconds(val):
    if isinstance(val, str) and ':' in val:
        mins, secs = val.split(':')
        return round(float(mins) * 60 + float(secs), 3)
    return val  # Leave as-is if not a proper time string

# Apply conversion
for col in ['q1', 'q2', 'q3']:
    qualifying[col] = qualifying[col].apply(to_seconds)
qualifying.head(20)
```

```python
lap_times["milliseconds"]  = lap_times["milliseconds"]/1000
lap_times.rename(columns={"milliseconds":"lap_time"}, inplace=True)
lap_times = lap_times[["raceId","driverId","lap","position","lap_time"]]
lap_times.head(20)
```

```python
qualifying = qualifying[["raceId","driverId","constructorId","number","position","q1","q2","q3"]]
qualifying.head(20)
```

```python
lap_times.rename(columns={"position":"lap_position"}, inplace=True)
qualifying.rename(columns={"position":"quali_position"}, inplace=True)
lap_times_qaulifying = pd.merge(lap_times, qualifying, on=["raceId","driverId"], how="left")
lap_times_qaulifying.head(600)
```

```python
pit_stops = datasets["pit_stops"].copy()
pit_stops.head(20)
```

```python
pit_stops["seconds"] = pit_stops["milliseconds"]/1000
pit_stops.rename(columns={"seconds":"pit_stop_duration", "stop":"no_of_stops"}, inplace=True)

pit_stops = pit_stops[["raceId","driverId","lap","no_of_stops","pit_stop_duration"]]
pit_stops.head(20)
```

```python
l_q_p = pd.merge(lap_times_qaulifying, pit_stops, on=["raceId","driverId","lap"], how="left")
l_q_p.head(600)
```

```python
results_df = datasets["results"].copy()
results_df.head(20)
```

```python
results_df = results_df[["raceId","driverId","number","laps","milliseconds","statusId"]]
results_df.head(600)
```

```python
def to_seconds(val):
    try:
        val = int(val)
        val = val/1000
        return val

    except:
        return val
results_df = results_df[results_df["statusId"] == 1]
results_df["milliseconds"] = results_df["milliseconds"].apply(to_seconds)
results_df.rename(columns= {"milliseconds":"total_time","number":"result_position", "laps":"total_laps"}, inplace=True)
results_df.head(600)
```

```python
results_df.head(160)
```

```python
final_df = pd.merge(l_q_p,results_df, on=["raceId","driverId"], how="left")
```

```python
final_df.head(600)
```

```python
races_df = datasets["races"].copy()
races_df = races_df[["raceId","year","round","name"]]
races_df.rename(columns={"name":"grand_prix_name"}, inplace=True)
drives_df = datasets["drivers"].copy()
drives_df = drives_df[["driverId","driverRef"]]
```

```python
final_df_1 = pd.merge(final_df, races_df, on=["raceId"], how="left")
final_df_1 = pd.merge(final_df_1, drives_df, on=["driverId"], how="left")
final_df_1.head(80)
```
